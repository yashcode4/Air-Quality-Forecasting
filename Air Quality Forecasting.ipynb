## Handling the missing values

**Convert all -200 to NaN**

**Replace all NaN values with mean of that specific column**

```python
air_quality_data = air_quality_data.replace(to_replace=-200, value=np.NaN)
```

```python
# Checking the number of missing values in the dataframe
air_quality_data.isnull().sum()
```

**Ths shows the actual number of missing values**

```python
air_quality_data.tail()
```

```python
# Select only columns with number data type
numeric_columns = air_quality_data.select_dtypes(include=['number'])

# Apply mean() function to those columns
numeric_columns.mean()
```

```python
# Replacing the missing values with mean value
# Select numeric columns only
numeric_columns = air_quality_data.select_dtypes(include='number').columns

# Fill missing values in numeric columns with their respective mean
air_quality_data[numeric_columns] = air_quality_data[numeric_columns].fillna(air_quality_data[numeric_columns].mean())
```

## Forecasting with FB Prophet Algorithm

FB Prophet Documentation: https://facebook.github.io/prophet/docs/quick_start.html 

```python
# Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'], format='mixed')
date_info
```

```python
# Converting the time from HH.MM.SS to HH-MM-SS
time_info = air_quality_data['Time']
time_info = time_info.apply(lambda x: x.replace('.','-'))
time_info
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# Combining 2 series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
date_time
```

```python
date_time.shape
```

```python
# Combining date and time
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```

```python
date_time
```

```python
date_time.info()
```

## We have to convert ds from object datatype to date time format

```python
# Empty Dataframe
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
data['ds'] = data['ds'].dt.tz_localize(None)
```

```python
data.head()
```

```python
data['y'] = air_quality_data['RH']
```

```python
data.head()
```

```python
# !pip install prophet
```

```python
from prophet import Prophet
```

```python
# Training the Prophet model
model = Prophet()
model.fit(data)
```

```python
# We have hourly data, that's why mentioning freq='H'. So it wil give forecasting for hourly data.
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plotting the forecast
fig1 = model.plot(forecast)
```

```python
# Plotting forecast components
fig2 = model.plot_components(forecast)
```
